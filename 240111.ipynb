```python
#한 화면에 여러 개 그래프 나타내기 : subplots( )
#subplots(행 개수, 열 개수, figsize=(가로 길이, 세로 길이)
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
fig, axs = plt.subplots(2,2) #axs행렬을 받아와서 저장하는 객체 
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

2.1 시본(Seaborn)이란?
1) 맷플롯립을 바탕으로 통계 그래프를 그릴 수 있도록 기능을 추가한 파이썬 라이브러리
2) 판다스의 데이터프레임을 대상으로 그래프를 그리기 때문에 데이터프레임을 다양한 그래프로 나타내기가 편리함
3) 타이타닉(titanic), 붓꽃(iris), 팁(tips), 여객운송(flights), 펭귄(penguins) 등의 샘플 데이터 셋을 제공
4) 시본이 설치되어 있지 않으면 아래의 명령을 실행해서 설치

```python
#산점도 그래프:scatterplot()
#scatterplot( x=‘x축에 표시할 칼럼명＇, y=‘y축에 표시할 칼럼명’,hue=‘데이터 구분 칼럼명‘, data=데이터프레임 )
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris") #seborn에서 제공해주는 sample 예제 데이터 셋
sns.scatterplot(x='petal_length',y='petal_width',hue='species',data=iris)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

```python
#범주별 데이터 개수를 표시하는 막대 그래프 : countplot( )
#countplot( x=‘x축에 범주로 표시할 칼럼명＇, hue=‘데이터 구분 칼럼명‘,data=데이터프레임 )
iris_data=sns.load_dataset("iris")
sns.countplot(x='species',data=iris_data)
plt.show()
```

```python
#범주별 항목값 평균을 표시하는 막대 그래프 : barplot( )
#barplot( x=‘x축에 표시할 칼럼명＇, y=‘y축에 표시할 칼럼명’, data=데이터프레임 hue=data )
import matplotlib.pyplot as plt
import seaborn as sns
iris_data=sns.load_dataset('iris')
sns.barplot(x='species',y='petal_length',data=iris_data, errorbar='sd')
plt.show()
```

```python
#숫자값을 갖는 칼럼들의 상관관계를 보여주는 그래프 : pairplot( )
#pairplot(데이터프레임, hue=‘범주를 갖는 칼럼명＇)
import matplotlib.pyplot as plt
import seaborn as sns
iris_data=sns.load_dataset('iris')
sns.pairplot(iris_data,hue='species')
plt.show()
```

```python
# 산점도와 추세선(회귀선)을 동시에 보여주는 그래프 : regplot( )
# regplot(x=‘x축에 표시할 칼럼명‘, y=‘y축에 표시할 칼럼명‘, 데이터프레임)
import matplotlib.pyplot as plt
import seaborn as sns
iris_data=sns.load_dataset('iris')
sns.regplot(x='petal_length',y='petal_width', data=iris_data)
plt.show()
```

```python
# 집계한 값에 따라 색깔을 표시하는 그래프 : heatmap( )
#• heatmap(data=2차원 행렬 형태로 전달된 데이터, annot=True/False,fmt=‘파이썬 서식 문자열‘, cbar=True/False)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
iris_data=sns.load_dataset('iris')
df = iris_data.corr(numeric_only=True)
sns.heatmap(data=df,annot=True,fmt='.2f',cbar=False)
plt.show()
```

```python
# MISSION 3
#시본에서 제공하는 데이터인 ‘penguins’를 이용해서 그래프 그리기

#1.penguins 데이터를 가져온 후 펭귄 품종(species)별 몸무게(body_mass_g)의 평균을 그래프로 그리는데, 아래 그림과 같이 성별을 구분해서 그려보자
import seaborn as sns
import pandas as pd
penguins=sns.load_dataset("penguins")
fig, ax = plt.subplots()
sns.barplot(x='species',y='body_mass_g',hue='sex',data=penguins,ax=ax)
ax.legend(loc='upper left',title='sex')


plt.show()
```

```python
#2. penguins 데이터의 숫자값을 갖는 항목들끼리의 상관관계 그래프를 그려보자.

sns.pairplot(penguins)
plt.show()
```

```python
#3.2번 미션의 결과에서 칼럼들간의 상관관계를 확인하고 관계가 있는 두 칼럼을 선택하여 산점도와 추세선이 함께 나타나는 그래프를 그려보자.
# 'flipper_length_mm'와 'body_mass_g' 간의 산점도와 추세선 그리기

sns.scatterplot(x='flipper_length_mm', y='body_mass_g', data=penguins, color='b', label='Data Points')
sns.regplot(x='flipper_length_mm', y='body_mass_g', data=penguins, scatter=False, color='r', label='Trend Line')
plt.legend()
plt.show()
```
